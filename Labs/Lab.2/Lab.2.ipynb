Requirements:

- Only allow placing a mark in a previously empty location.
- Return `True` if the move was successful, and `False` otherwise.


````python
def make_move(board, player, row, col):
    # player 1 = "X", player 2 = "O"
    if player == 1:
        mark = "X"
    else:
        mark = "O"

    n = len(board)
    if row < 0 or row >= n or col < 0 or col >= n:
        return False

    # only place on empty spot
    if board[row][col] == "":
        board[row][col] = mark
        return True
    else:
        return False
````

*Exercise 9:* Use all of the previous exercises to implement a full tic‑tac‑toe game:

- draw the board,
- repeatedly ask two players for a location,
- apply valid moves,
- check the game status until a player wins or the game is a draw.


````python
def print_board_with_labels(board):
    n = len(board)
    letters = [chr(ord("A") + i) for i in range(n)]
    print("  " + " ".join(letters))

    for i in range(n):
        row_label = str(i + 1)
        row = board[i]
        row_str = " ".join(cell if cell != "" else "." for cell in row)
        print(row_label + " " + row_str)


# 5

def make_move(board, player, row, col):
    if player == 1:
        mark = "X"
    else:
        mark = "O"

    n = len(board)
    if row < 0 or row >= n or col < 0 or col >= n:
        return False

    if board[row][col] == "":
        board[row][col] = mark
        return True
    else:
        return False


# 6

def parse_move(pos):
    pos = pos.strip().upper()
    if len(pos) != 2:
        return None

    col_char = pos[0]
    row_char = pos[1]

    if not col_char.isalpha() or not row_char.isdigit():
        return None

    col = ord(col_char) - ord("A")
    row = int(row_char) - 1

    return (row, col)


# 7

def make_move_from_location(board, player, location):
    result = parse_move(location)
    if result is None:
        return False

    row, col = result
    return make_move(board, player, row, col)


# 8

def get_player_move(board, player):
    while True:
        print_board_with_labels(board)
        move = input("Player " + str(player) + " (" + ("X" if player == 1 else "O") +
                     ") enter your move (like A2): ")

        if make_move_from_location(board, player, move):
            print("Move accepted.\n")
            break
        else:
            print("Invalid move, try again.\n")


# 4

def tic_tac_toe_state(board):
    n = len(board)
    empty_found = False

    for i in range(n):
        row = board[i]
        if "" in row:
            empty_found = True
        if row.count("X") == n:
            return 1
        if row.count("O") == n:
            return 2

    for j in range(n):
        col = []
        for i in range(n):
            col.append(board[i][j])
        if col.count("X") == n:
            return 1
        if col.count("O") == n:
            return 2

    main_diag = []
    for i in range(n):
        main_diag.append(board[i][i])
    if main_diag.count("X") == n:
        return 1
    if main_diag.count("O") == n:
        return 2

    anti_diag = []
    for i in range(n):
        anti_diag.append(board[i][n - 1 - i])
    if anti_diag.count("X") == n:
        return 1
    if anti_diag.count("O") == n:
        return 2

    # no winner
    if empty_found:
        return -1   
    else:
        return 0  


# 9

def play_game():

    board = [
        ["", "", ""],
        ["", "", ""],
        ["", "", ""]
    ]

    current_player = 1 

    while True:
        get_player_move(board, current_player)

        print_board_with_labels(board)
        state = tic_tac_toe_state(board)

        if state == 1:
            print("Player 1 (X) wins!")
            break
        elif state == 2:
            print("Player 2 (O) wins!")
            break
        elif state == 0:
            print("Game is a draw.")
            break
        else:
            if current_player == 1:
                current_player = 2
            else:
                current_player = 1
            print() 
````

*Exercise 11:* Develop a version of the game where one player is the computer.

Note: you do **not** need an extensive search for the best move. For example, you can have the computer:
- block obvious losses
- otherwise try to create a winning row/column/diagonal


````python
import random

def computer_move(board, computer="O", player="X"):
    n = len(board)

    def find_winning_move(line):
        if line.count(computer) == n - 1 and line.count("") == 1:
            return line.index("")
        if line.count(player) == n - 1 and line.count("") == 1:
            return line.index("")
        return -1

    for i in range(n):
        idx = find_winning_move(board[i])
        if idx != -1:
            board[i][idx] = computer
            return

    for j in range(n):
        col = [board[i][j] for i in range(n)]
        idx = find_winning_move(col)
        if idx != -1:
            board[idx][j] = computer
            return

    diag = [board[i][i] for i in range(n)]
    idx = find_winning_move(diag)
    if idx != -1:
        board[idx][idx] = computer
        return

    anti_diag = [board[i][n - 1 - i] for i in range(n)]
    idx = find_winning_move(anti_diag)
    if idx != -1:
        board[idx][n - 1 - idx] = computer
        return

    empty_cells = [(i, j) for i in range(n) for j in range(n) if board[i][j] == ""]
    if empty_cells:
        i, j = random.choice(empty_cells)
        board[i][j] = computer
````

````python
board = [
    ["X", "O", "X"],
    ["", "X", ""],
    ["", "", "O"]
]

computer_move(board, computer="O", player="X")

for row in board:
    print(row)
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
import copy

def minimax(board, depth, is_maximizing, max_depth, computer="O", player="X"):
    state = tic_tac_toe_state(board)
    
    if state == 1:
        return -1 if computer == "O" else 1
    if state == 2:
        return 1 if computer == "O" else -1
    if state == 0 or depth == max_depth:
        return 0

    n = len(board)

    if is_maximizing:
        best_score = -float("inf")
        for i in range(n):
            for j in range(n):
                if board[i][j] == "":
                    board[i][j] = computer
                    score = minimax(board, depth + 1, False, max_depth, computer, player)
                    board[i][j] = ""
                    best_score = max(best_score, score)
        return best_score
    else:
        best_score = float("inf")
        for i in range(n):
            for j in range(n):
                if board[i][j] == "":
                    board[i][j] = player
                    score = minimax(board, depth + 1, True, max_depth, computer, player)
                    board[i][j] = ""
                    best_score = min(best_score, score)
        return best_score

def computer_move_minimax(board, max_depth=3, computer="O", player="X"):
    n = len(board)
    best_score = -float("inf")
    move = None

    for i in range(n):
        for j in range(n):
            if board[i][j] == "":
                board[i][j] = computer
                score = minimax(board, 0, False, max_depth, computer, player)
                board[i][j] = ""
                if score > best_score:
                    best_score = score
                    move = (i, j)
    
    if move:
        board[move[0]][move[1]] = computer
````

````python
board = [
    ["X", "", ""],
    ["", "O", ""],
    ["", "", "X"]
]

computer_move_minimax(board, max_depth=3, computer="O", player="X")

for row in board:
    print(row)
````

*Exercise 13:* Make the 2 computer players play each-other for 10 games on a 3x3, then 4x4, then 5x5 grid. Set the max depth so that the games only take seconds. Measure the "smarter" player's win rate for each grid.

````python
import random

def play_game(n, depth_X, depth_O):
    board = [["" for _ in range(n)] for _ in range(n)]
    current = "X" 
    
    while True:
        state = tic_tac_toe_state(board)
        if state != -1: 
            return state  

        if current == "X":
            computer_move_minimax(board, max_depth=depth_X, computer="X", player="O")
            current = "O"
        else:
            computer_move_minimax(board, max_depth=depth_O, computer="O", player="X")
            current = "X"

def run_simulation():
    sizes = [3, 4, 5] 
    results = {} 
    
    for n in sizes:
        wins = {"smarter": 0, "weaker": 0, "draw": 0}
        if n == 3:
            smarter = 3
        else:
            smarter = 2 
        weaker = 1
        
        for i in range(10):  
            result = play_game(n, smarter, weaker)
            if result == 1:
                wins["smarter"] += 1
            elif result == 2:
                wins["weaker"] += 1
            else:
                wins["draw"] += 1
        
        results[n] = wins
    
    for n in results:
        print(f"\nBoard size: {n}x{n}")
        print(f"Smarter player wins: {results[n]['smarter']}")
        print(f"Weaker player wins: {results[n]['weaker']}")
        print(f"Draws: {results[n]['draw']}")

run_simulation()
````
