1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_value):
        self.value = 0
        self.max_value = max_value

    def increment(self):
        if self.value < self.max_value:
            self.value += 1
        else:
            print("Error: Cannot go above the maximum value!")

    def reset(self):
        self.value = 0

    def get_value(self):
        return self.value

c = Counter(5)

for i in range(7):
    print("Counter is at:", c.get_value())
    c.increment()

print("Resetting counter...")
c.reset()
print("Counter is now:", c.get_value())
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
class GameObject:
    def __init__(self, x, y, width, height):
        self.x = x
        self.y = y
        self.width = width
        self.height = height

    def overlaps(self, other):
        if (self.x < other.x + other.width and
            self.x + self.width > other.x and
            self.y < other.y + other.height and
            self.y + self.height > other.y):
            return True
        else:
            return False


class Player(GameObject):
    pass

class Enemy(GameObject):
    pass

p = Player(10, 10, 20, 20)
e = Enemy(25, 25, 20, 20)

print(p.overlaps(e))
```

```python
import math
import tkinter as tk

class Canvas:
    def __init__(self , w=400, h=400):
        self.root = tk.Tk()
        self.canvas = tk.Canvas(self.root , width=w , height=h)
        self.canvas.pack()
        self.w = w
        self.h = h

    def draw_line(self , x1,y1,x2,y2 , color="black"):
        self.canvas.create_line(x1,y1,x2,y2,fill=color)

    def draw_circle(self , x,y,r , color="black"):
        self.canvas.create_oval(x-r , y-r , x+r , y+r, outline=color)

    def draw_rect(self , x,y,w,h , color="black"):
        self.canvas.create_rectangle(x,y,x+w,y+h, outline=color)

    def show(self):
        self.root.mainloop()


class Shape:
    def __init__(self):
        pass

    def area(self):
        raise NotImplementedError()

    def perimeter(self):
        raise NotImplementedError()

    def getPoints(self):
        raise NotImplementedError()

    def inside(self,x,y):
        raise NotImplementedError()

    def paint(self , canvas):
        raise NotImplementedError()

    def overlaps(self, other):
        pts = self.getPoints()
        for (xx,yy) in pts:
            if other.inside(xx,yy):
                return True
        return False



class Rectangle(Shape):
    def __init__(self , length , width , x , y ):
        super().__init__()
        self.__l = length
        self.__w = width
        self.__x = x
        self.__y = y

    def area(self): return self.__l * self.__w
    def perimeter(self): return 2*(self.__l + self.__w)

    def inside(self,x,y):
        return (self.__x <= x <= self.__x+self.__l) and (self.__y <= y <= self.__y+self.__w)

    def getPoints(self):
        pts=[]
        for i in range(16):
            t = i/16
            pts.append((self.__x + self.__l*t , self.__y))
            pts.append((self.__x + self.__l , self.__y + self.__w*t))
        return pts[:16]

    def paint(self,canvas):
        canvas.draw_rect(self.__x , self.__y , self.__l , self.__w)



class Circle(Shape):
    def __init__(self , r , x , y ):
        super().__init__()
        self.__r = r
        self.__x = x
        self.__y = y

    def area(self): return math.pi*(self.__r**2)
    def perimeter(self): return 2*math.pi*self.__r

    def inside(self,x,y):
        dx = x - self.__x
        dy = y - self.__y
        return dx*dx + dy*dy <= self.__r*self.__r

    def getPoints(self):
        pts=[]
        for i in range(16):
            a = 2*math.pi * (i/16)
            xx = self.__x + self.__r*math.cos(a)
            yy = self.__y + self.__r*math.sin(a)
            pts.append((xx,yy))
        return pts

    def paint(self , canvas):
        canvas.draw_circle(self.__x , self.__y , self.__r)


class Triangle(Shape):
    def __init__(self , ax,ay , bx,by , cx,cy ):
        super().__init__()
        self.ax=ax; self.ay=ay
        self.bx=bx; self.by=by
        self.cx=cx; self.cy=cy

    def area(self):
        return abs(self.ax*(self.by-self.cy)+self.bx*(self.cy-self.ay)+self.cx*(self.ay-self.by))/2

    def perimeter(self):
        d=lambda x1,y1,x2,y2: math.sqrt((x2-x1)**2 + (y2-y1)**2)
        return d(self.ax,self.ay,self.bx,self.by)+d(self.bx,self.by,self.cx,self.cy)+d(self.cx,self.cy,self.ax,self.ay)

    def inside(self,x,y):
        A=self.area()
        A1=Triangle(x,y, self.bx,self.by, self.cx,self.cy).area()
        A2=Triangle(self.ax,self.ay, x,y, self.cx,self.cy).area()
        A3=Triangle(self.ax,self.ay, self.bx,self.by, x,y).area()
        return abs(A - (A1+A2+A3)) < 0.01

    def getPoints(self):
        pts=[]
        corners=[(self.ax,self.ay), (self.bx,self.by), (self.cx,self.cy)]
        for i in range(3):
            x1,y1=corners[i]
            x2,y2=corners[(i+1)%3]
            for t in range(6):
                u=t/6
                pts.append((x1+(x2-x1)*u , y1+(y2-y1)*u))
        return pts[:16]

    def paint(self,canvas):
        canvas.draw_line(self.ax,self.ay,self.bx,self.by)
        canvas.draw_line(self.bx,self.by,self.cx,self.cy)
        canvas.draw_line(self.cx,self.cy,self.ax,self.ay)

class CompoundShape(Shape):
    def __init__(self):
        self.shapes=[]

    def add(self , s):
        self.shapes.append(s)

    def area(self):
        return sum([s.area() for s in self.shapes])

    def perimeter(self):
        return sum([s.perimeter() for s in self.shapes])

    def inside(self,x,y):
        for s in self.shapes:
            if s.inside(x,y):
                return True
        return False

    def getPoints(self):
        allp=[]
        for s in self.shapes:
            allp += s.getPoints()
        return allp[:16]

    def paint(self , canvas):
        for s in self.shapes:
            s.paint(canvas)


if __name__ == "__main__":
    c = Canvas(500,500)

    r = Rectangle(150,80 , 50,50)
    circle = Circle(60 , 300 , 200)
    tri = Triangle(200,300 , 250,150 , 350,300)

    comp = CompoundShape()
    comp.add(Rectangle(60,40, 100,300))
    comp.add(Circle(30,160,330))

    r.paint(c)
    circle.paint(c)
    tri.paint(c)
    comp.paint(c)

    c.show()
```

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```
